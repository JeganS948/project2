```python
# # drop columns
columns=["Departure", "Return", "Bike", "Departure station", "Return station", "Membership type", "Duration (sec.)"]
vancouver_01 = vancouver_01[columns]
# vancouver_02= vancouver_02.rename(columns={"Membership Type": "Membership type"})
# vancouver_02 = vancouver_02[columns]
# vancouver_03 = vancouver_03[columns]
# vancouver_04 = vancouver_04[columns]
# vancouver_05 = vancouver_05[columns]
# vancouver_06 = vancouver_06[columns]
# vancouver_07 = vancouver_07[columns]
# vancouver_08 = vancouver_08[columns]
# vancouver_09 = vancouver_09[columns]
# vancouver_10 = vancouver_10[columns]
# vancouver_11 = vancouver_11[columns]
# vancouver_12= vancouver_12.rename(columns={"Membership Type": "Membership type"})
# vancouver_12 = vancouver_12[columns]


```

```python
#performing concatenation
# vancouver = [vancouver_01,vancouver_02,vancouver_03,vancouver_04,vancouver_05,vancouver_06,
#              vancouver_07,vancouver_08,vancouver_09,vancouver_10,vancouver_11,vancouver_12]
# bike_share = pd.concat(vancouver).drop_duplicates().reset_index(drop=True)
bike_share = vancouver_01
```

```python
# remove rows with null values
bike_share = bike_share.dropna()
```

```python
#drop bad dates
bike_share= bike_share.loc[(bike_share["Return"]!= "1900-01-00 0:00")]
bike_share
```

```python
#Add new column:location_id
bike_share = bike_share.assign(location_id=2)
```

```python
bike_share.head()
```

```python
#Renaming columns
bike_share = bike_share.rename(columns={"Bike": "bike_id",
                                        "Departure station":"start_station_id",
                                        "Return station":"end_station_id",
                                        "Departure":"start_date",
                                        "Return":"end_date",
                                        "Duration (sec.)":"trip_duration",
                                        "Membership type":"member_type"})

bike_share
```

```python
#Getting Departure station id
bike_share["start_station_id"]= bike_share["start_station_id"].str.split(" ", n = 1, expand = True) 
```

```python
# Getting Return station id
bike_share["end_station_id"]= bike_share["end_station_id"].str.split(" ", n = 1, expand = True) 
```

```python
#Rearrange columns
bike_share = bike_share[["trip_duration","start_date","end_date","start_station_id","end_station_id",
                         "bike_id","member_type","location_id"]]
bike_share
```

```python
#total count of list
bike_share.count()
```

```python
#Bike ID
bikes = pd.DataFrame(bike_share[["bike_id","location_id"]])
bikes = bikes.drop_duplicates().reset_index(drop=True)
```

```python
bikes
```

```python
#Membership :User type
usertype = pd.DataFrame(bike_share[["member_type","location_id"]])
usertype = usertype.drop_duplicates().reset_index(drop=True)
usertype
```
