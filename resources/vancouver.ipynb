```python
# import dependencies
import requests
import json
import pandas as pd
import numpy as np
```

```python
# URL for Get requets to retireve data 
url = "https://vancouver-gbfs.smoove.pro/gbfs/en/station_information.json"

# URL's for other cities station information
#toronto url = "https://toronto-us.publicbikesystem.net/ube/gbfs/v1/en/station_information"
#new york url = "https://gbfs.citibikenyc.com/gbfs/en/station_information.json"
#boston url = "https://gbfs.bluebikes.com/gbfs/en/station_information.json"
```

```python
# print the response object to the console
print(requests.get(url))
```

```python
# print JSON for all launchpads
x = requests.get(url).json()
print(json.dumps(x, indent=4, sort_keys=True))
```

```python
# create a list of stations
stations = x["data"]["stations"]
```

```python
# Lists to hold response data 
station_id = []
name = []
lat = []
lon = []

# loop and make requests and parse
for i in stations:
#target_url = url + stations
#response = requests.get(target_url).json()
    station_id.append(i["station_id"])
    name.append(i["name"])
    lat.append(i["lat"])
    lon.append(i["lon"])    
```

```python
# build a dataframe
station_info = pd.DataFrame({
    "station_id": station_id,
    "name": name,
    "lat": lat,
    "lon": lon})
station_info
```
